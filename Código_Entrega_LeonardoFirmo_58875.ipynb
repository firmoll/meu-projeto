```python
#Carregando bibliotecas
import pandas as pd
import requests
import pprint
import sqlite3
from plyer import notification 
```

Carregar as APIs

```python
#Carregando a primeira API a ser estudada - Bancos
url = "https://brasilapi.com.br/api/banks/v1"
API1 = requests.get(url)

#Checando a requisição de chamada da primeira API
if API1.status_code == 200:
    data_json = API1.json()
else:
    print("Erro ao acessar a API 1")

#Carregando a segunda API - Corretoras
url = "https://brasilapi.com.br/api/cvm/corretoras/v1"
API2 = requests.get(url)

#Checando a requisição de chamada da segunda API
if API2.status_code == 200:
    data2_json = API2.json()
else:
    print("Erro ao acessar a API 2")

#Carregando a última API a ser estudada - - Códigos NCM / Receita Federal
url = "https://brasilapi.com.br/api/ncm/v1"
API3 = requests.get(url)

#Checando a requisição de chamada da última API
if API3.status_code == 200:
    data3_json = API3.json()
else:
    print("Erro ao acessar a API 3")
```

Visualizar os dados das APIs

```python
#Carregando os dados - Lista de Bancos
data_json

#Tratando a API
Bancos=pd.DataFrame(data_json)
Bancos.head(27)
```

```python
#Carregando os dados - Lista de Corretoras
data2_json

#Tratando os dados
Corretoras=pd.DataFrame(data2_json)
Corretoras.head(27)
```

```python
#Carregando os dados - Códigos NCM / Receita Federal
data3_json

#Tratando os dados
NCM=pd.DataFrame(data3_json)
NCM.head(27)
```

Exportar os dados para uma tabela CSV

```python
# Exportar os dados desta API para um arquivo CSV

#Bancos
df1 = pd.json_normalize(data_json)
df1.to_csv('Bancos.csv', index=False)

#Check
df2 = pd.json_normalize(data2_json)
df2.to_csv('Corretoras.csv', index=False)

#Area
df3 = pd.json_normalize(data3_json)
df3.to_csv('NCMs.csv', index=False)
```

Conectar e criar um banco de dados

```python
# Conectar ao banco de dados
conn = sqlite3.connect('CursoPython_Turma58875.db')

# Escrever o DataFrame nas tabelas
Bancos.to_sql('Bancos',conn,if_exists='replace',index=False)
Corretoras.to_sql('Corretoras',conn,if_exists='replace',index=False)
NCM.to_sql('Area',conn,if_exists='replace',index=False)

# Confirmar as inserções
conn.commit()

# Fechar conexão
conn.close()
```
